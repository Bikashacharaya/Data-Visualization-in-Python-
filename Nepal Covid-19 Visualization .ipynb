```python
import matplotlib.pyplot as plt
import numpy as np
%matplotlib inline 
```

```python
company=['Google','Infosis','Amazon','Facebook']
revenue=[80,70,50,60]
profit=[50,35,10,45]
```

```python
gog=np.arange(len(company))
gog
```

```python
plt.xticks(gog,company)
plt.bar(gog-0.2,revenue, width=0.4, label="Revenue")
plt.bar(gog+0.2,profit, width=0.4, label="Profit")
plt.legend()
plt.show()
```

```python
days=[1,2,3,4,5,6,7]
max=[31,32,36,42,45,50,30]
min=[21,25,26,22,19,18,16]
avg=[26,29,25,24,30,27,30]
```

```python
plt.plot(days,max, label="MAX")
plt.plot(days,min, label="MIN")
plt.plot(days,avg, label="AVG")
plt.legend()
plt.grid()
```

```python
import numpy as np
import matplotlib.pyplot as plt
%matplotlib inline 
```

```python

date=["1-july","2-july","3-july","4-july","5-july","6-july","7-july","8-july","9-july","10-july","11-july","12-july"]
newcases=[482,473,470,232,292,180,204,255,108,118,70,82]
recovered=[1462,664,823,272,132,264,688,253,139,120,431,147]
death=[1,1,1,2,0,1,0,0,0,0,3,0]
```

```python
dt=np.arange(len(date))
```

```python
plt.figure(figsize=(15,5))
plt.xticks(dt,date)
plt.plot(dt,newcases, label="New-case")
plt.plot(dt,recovered, label="Recovered")
plt.plot(dt,death , label="Death")
plt.xlabel("Date")
plt.ylabel("Frequency")
plt.title("Nepal-Covid Dataset Month of July ")
plt.legend()
plt.grid()
```
